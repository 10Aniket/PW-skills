```python
def print_exception_hierarchy():
    print("Python Exception Hierarchy:\n")

    # Iterate over the exception hierarchy and print each class name
    for exc in Exception.__subclasses__():
        print(exc.__name__)

        # Iterate over the subclasses of the current exception class
        for subexc in exc.__subclasses__():
            print(f"  {subexc.__name__}")

        print()

print_exception_hierarchy()
```

1.Be specific with your exceptions: Use specific exception classes to catch only the errors that you expect. For example,
    use FileNotFoundError to catch file not found errors instead of using a generic Exception class.

2.Use the finally block for cleanup: Use the finally block to clean up resources like files, sockets, and database connections,
    regardless of whether an exception is raised or not.

3.Keep your try blocks small: Put only the code that might raise an exception inside a try block.
    This makes it easier to identify which code is causing the exception.
